# Singly Linked List - Deletion

```python
class Node():
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList():
    def __init__(self):
        self.head = None

    def printList(self):
        temp = self.head
        while temp:
            print(temp.data,end = ' --> ')
            temp = temp.next
        print("None")

    def push(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        new_node.next = self.head
        self.head = new_node

    def addAfter(self,prev_node,data):
        new_node = Node(data)
        if prev_node is None:
            print("The given node must be in LinkedList")
            return
        new_node.next = prev_node.next
        prev_node.next = new_node

    def append(self,data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    def deleteNode(self,key):
        # deletes the node with first occurence of given key
        
        temp = self.head
        # if head node contains key
        if temp is not None:
            if temp.data == key:

                self.head = temp.next
                temp = None
                return
        # we search for the key, while traversing through the linkedlist
        while temp is not None:
            if temp.data == key:
                break
            prev = temp
            temp = temp.next

        # if key is not present in the LinkedList
        if temp == None:
            print("The given key is not in LinkedList")
            return

        prev.next = temp.next
        temp = None
        
    def deleteNode_atPosition(self,pos):
        # deletes the node at given position 
        # if LinkedList is empty
        if self.head is None:
            print("The LinkedList is empty")
            return
        
        temp = self.head
        # if position is 0 - we delete first node
        if pos == 0:
            self.head = temp.next
            temp = None
            return
        prev = None
        count = 1
        while temp and count <= pos:
            prev = temp
            temp = temp.next
            count += 1
        if temp is None:
            print("The given position is out of range of LinkedList index")
            return
        prev.next = temp.next
        temp = None
```

```python
llist = LinkedList()
print("Empty list:")
llist.printList()
print("\n\nList with 3 nodes:")
llist.push(20)
llist.push(10)
llist.push(5)

llist.printList()

llist.addAfter(llist.head.next,15)
print("\n\nElement added after 10:")
llist.printList()
print("\n\nElemnts added at the end:")
llist.append(25)
llist.append(30)

llist.printList()

llist.deleteNode(15)
print("\n\nDeleting element 15:")
llist.printList()

llist.deleteNode_atPosition(3)
print("\n\nDeleting node at position 3")
llist.printList()
```
